```python
# create a csv file of the filename (optionally without file extension)
def generate_files_csv(filename='filenames.csv', save_dir="./tmp", file_dir=os.path.join("./tmp","patches/"), 
                       cats=[], drop_new=True, drop_ext=True):
    # filename: name of csv containing files
    # save_dir: directory to save csv
    # file_dir: directory to read the files from. The names of files are used to generate each entry
    # cats: array of categories to include only these types. If empty array, or None, include everything
    # drop_new: if the category name contains 'new' (e.g. AdCC new), ignore them
    
    # get files without extensions
    filenames = [f for f in os.listdir(file_dir) if os.path.isfile(os.path.join(file_dir, f))]
    if drop_ext:
        filenames = [os.path.splitext(f)[0] for f in filenames]

    df = pd.DataFrame(filenames, columns=['slide_id'])
    # df = df.drop(df["slide_id"<some boolean condition>].index)
    if cats == [] or None:
        pass
    else:
        df = df[df.slide_id.str.contains('|'.join(cats), case=False)]

    if drop_new:
        df = df[~df.slide_id.str.contains('New', case=False)]
        df = df.reset_index(drop=True)
    
    # write to a csv
    print(f"saving dataframe to csv {os.path.join(save_dir,'filenames.csv')}")
    df.to_csv(os.path.join(save_dir,'filenames.csv'), index = False)
    
    return df
```

### Patch extraction - patch coords

```python
def get_patches_coords(resolution, units, wsi, mask, min_mask_ratio):
    wsi_shape = wsi.slide_dimensions(resolution=resolution, units=units)

    (patch_inputs, _) = tiatoolbox.tools.patchextraction.PatchExtractor.get_coordinates(
        image_shape=wsi_shape,
        patch_input_shape=rect_size,
        patch_output_shape=rect_size,
        stride_shape=[224,224],
    )

    # Filter out coords which dont lie in mask
    selected_coord_indices = tiatoolbox.tools.patchextraction.PatchExtractor.filter_coordinates(
        WSIReader.open(mask),
        patch_inputs,
        wsi_shape=wsi_shape,
        min_mask_ratio=min_mask_ratio,
    )
    patch_inputs = patch_inputs[selected_coord_indices]
    return patch_inputs
```

## Patch extraction for every image

```python
class SingleWSIDataset(torch.utils.data.Dataset):
    def __init__(self, slide_name, wsi_img_path=data_dir, patch_csv='./tmp/patches'):
        self.slide_name = slide_name
        self.patch_df = pd.read_csv(os.path.join(patch_csv,f"{slide_name}.csv"), header=None)
        self.wsi = WSIReader.open(input_img=os.path.join(wsi_img_path,f"{slide_name}.svs"))
        
    def __len__(self):
        return len(self.patch_df)

    def __getitem__(self, idx):
#         img_path = os.path.join(self.img_dir, self.img_labels.iloc[idx, 0])
#         image = read_image(img_path)
#         label = self.img_labels.iloc[idx, 1]
#         if self.transform:
#             image = self.transform(image)
#         if self.target_transform:
#             label = self.target_transform(label)
        patch_coords = self.patch_df.iloc[[idx]].to_numpy()[0]

        image = wsi.read_rect(
            location=(patch_coords[0], patch_coords[1]),
            size=(patch_coords[2]-patch_coords[0], patch_coords[3]-patch_coords[1]),
            resolution=1,
            units="mpp",
            coord_space="resolution"
        )
        # save patch coordinates as the label
        label = patch_coords

        return image, label
```

```python
def save_patches_csv(df, data_dir=data_dir, save_dir='./tmp/patches', skip_existing=True):
    # generate patches and generate a mask. Then save results
    # will overwrite df
    
    patch_resolution = 1
    patch_units = "mpp"
    mask_resolution=1.25
    mask_units="power"
    
    new_df = df.copy()
    
    for i, slide in enumerate(tqdm(df["slide_id"])):
        try:
            wsi = WSIReader.open(input_img=os.path.join(data_dir,slide))
        except:
            print(f"issue with {slide}")
            continue
        mask = wsi.tissue_mask(resolution=mask_resolution, units=mask_units)
        coords = get_patches_coords(patch_resolution, patch_units, wsi, mask, 0.7)
        
        num_patches = len(coords)
        if num_patches == 0:
            continue
        new_df.at[i,"patches"] = int(num_patches)
        
        # save to a file for easy referecing
        filename_no_ext = os.path.splitext(slide)[0]
        df_coords = pd.DataFrame(coords)
        filename = os.path.join(save_dir,f"{filename_no_ext}.csv")
        df_coords.to_csv(filename, index=False, header=False)
#         save_patches_to_csv_file(os.path.join(save_dir,f"{filename_no_ext}.csv"))
#         df = pd.DataFrame(coords)
        if os.path.exists(filename) and skip_existing:
            continue
        else:
            df_coords.to_csv(filename, index=False, header=False)
    new_df = new_df.dropna()
    new_df = new_df.reset_index(drop=True)
    return new_df
```

```python
import timm
from timm.models.layers.helpers import to_2tuple
import torch.nn as nn
```

download ctranspath weights as shown in https://github.com/Xiyue-Wang/TransPath

Also follow instructions for installing timm from archived file

```python
def load_ctranspath_model(load_model_path = './models/ctranspath.pth'):
    # from the ctranspath repo https://github.com/Xiyue-Wang/TransPath
    class ConvStem(nn.Module):

        def __init__(self, img_size=224, patch_size=4, in_chans=3, embed_dim=768, norm_layer=None, flatten=True):
            super().__init__()

            assert patch_size == 4
            assert embed_dim % 8 == 0

            img_size = to_2tuple(img_size)
            patch_size = to_2tuple(patch_size)
            self.img_size = img_size
            self.patch_size = patch_size
            self.grid_size = (img_size[0] // patch_size[0], img_size[1] // patch_size[1])
            self.num_patches = self.grid_size[0] * self.grid_size[1]
            self.flatten = flatten


            stem = []
            input_dim, output_dim = 3, embed_dim // 8
            for l in range(2):
                stem.append(nn.Conv2d(input_dim, output_dim, kernel_size=3, stride=2, padding=1, bias=False))
                stem.append(nn.BatchNorm2d(output_dim))
                stem.append(nn.ReLU(inplace=True))
                input_dim = output_dim
                output_dim *= 2
            stem.append(nn.Conv2d(input_dim, embed_dim, kernel_size=1))
            self.proj = nn.Sequential(*stem)

            self.norm = norm_layer(embed_dim) if norm_layer else nn.Identity()

        def forward(self, x):
            B, C, H, W = x.shape
            assert H == self.img_size[0] and W == self.img_size[1], \
                f"Input image size ({H}*{W}) doesn't match model ({self.img_size[0]}*{self.img_size[1]})."
            x = self.proj(x)
            if self.flatten:
                x = x.flatten(2).transpose(1, 2)  # BCHW -> BNC
            x = self.norm(x)
            return x
    model = timm.create_model('swin_tiny_patch4_window7_224', embed_layer=ConvStem, pretrained=False)
    model.head = nn.Identity()
    td = torch.load(load_model_path)
    model.load_state_dict(td['model'], strict=True)
    return model
```

## Extract features for every image

```python
def generate_features(df, save_path="./tmp/features/", skip_existing=True):
    model_gpu.eval()
    with torch.no_grad():
        for index, row in tqdm(df.iterrows(), total=df.shape[0]):
            slide_name = os.path.splitext( row['slide_id'] )[0]
            save_name = os.path.join(save_path, f"{slide_name}.pt")
            
            if os.path.exists(save_name) and skip_existing:
                continue
            
            single_wsi_dataset = SingleWSIDataset(slide_name)
            dataloader = torch.utils.data.DataLoader(single_wsi_dataset, batch_size=64, shuffle=False)

            preds = torch.randn(0,768) # create a 0-length intial tensor
            print(f"Extracting features for {slide_name}")
            for batch_idx, (img_sec, coords) in enumerate(tqdm(dataloader, leave=False)):
                img_sec = img_sec.permute(0, 3, 1, 2)
                img_sec = img_sec.float().to(device_string)
                features = model_gpu(img_sec)
                features_cpu = features.to('cpu')
                preds = torch.cat((preds, features_cpu), dim=0)
                        
            torch.save(preds, save_name)
```
